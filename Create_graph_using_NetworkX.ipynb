```python
# Creating graph using NetworkX
import networkx as nx
import math
import random
import datetime
import matplotlib.pyplot as plt
seed=10         # seed the graph for dependableness
c=100
G= nx.gnp_random_graph (c,.3, seed=seed )
print ( G.nodes() )
nx.draw(G,with_labels=True)
```
